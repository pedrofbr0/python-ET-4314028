```python
#My code:
def encodeString(stringVal):
    charLast = stringVal[0]
    count = 0
    encoded = []

    for char in stringVal:
        if char != charLast:
            encoded.append((charLast,count))
            count = 0
            charLast = char
        count += 1
    else:
        encoded.append((charLast,count))   

    return encoded

def decodeString(encodedList):
    string = ''
    for char, count in encodedList:
        string += char * count
        
    return string

test = "Hello World"
encoded = encodeString(test)
print(encoded)
decoded = decodeString(encoded)
print(decoded)
```
